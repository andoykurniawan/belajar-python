```python
import numpy as np
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y);
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Judul Grafik')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.subplot(122)
plt.plot(y,x, color='blue')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.subplot(122)
plt.plot(y,x, color='blue')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.tight_layout()
plt.show()
```
