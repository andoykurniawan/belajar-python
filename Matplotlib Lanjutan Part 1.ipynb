```python
import pandas as pd

# dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/retail_raw_reduced.csv")
dataset = pd.read_csv("retail_raw_reduced.csv")
dataset
```

```python
# untuk menampilkan baris dan kolom nya
dataset.shape
```

```python
# info dataset
dataset.info()
```

```python
# pengubahan / transformasi data menjadi string
dataset[['order_id', 'customer_id']] = dataset[['order_id', 'customer_id']].astype('str')
```

```python
dataset.info()
```

```python
import datetime

# untuk mendapatkan bulan dan tahun pada kolom order date
dataset['order_month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
dataset
```

```python
dataset.info()
```

```python
# Penambahan kolom GMV (Gross Merchandise Value) pada dataset / Nilai total
dataset['gmv'] = dataset['quantity'] * dataset['item_price']
```

```python
dataset.info()
```

```python
# Membuat Data Agregat
monthly_amount = dataset.groupby('order_month')['gmv'].sum().reset_index()
monthly_amount
```

```python
# Membuat plot Line Chart Trend Pertumbuhan GMV
import matplotlib.pyplot as plt

plt.plot(monthly_amount['order_month'], monthly_amount['gmv'])
plt.show()
```

```python
# Cara Alternatif: Fungsi .plot() pada pandas Dataframe
dataset.groupby(['order_month'])['gmv'].sum().plot()
plt.show()
```

```python
# Mengubah Figure Size
plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot()
plt.show()
```

```python
# Menambahkan Title dan Axis Labels
dataset.groupby(['order_month'])['gmv'].sum().plot()
plt.title('Monthly GMV Year 2019')
plt.xlabel('Order Month')
plt.ylabel('Total GMV')
plt.show()
```

```python
# Kustomisasi Title dan Axis
dataset.groupby(['order_month'])['gmv'].sum().plot()
plt.title('Monthly GMV Year 2019', loc='center', fontsize = 20, color = 'blue', pad = 20)
plt.xlabel('Order Month',fontsize = 15)
plt.ylabel('Total GMV', fontsize =  15)
plt.show()
```

```python
# Kustomisasi grafik plot
plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color = 'green', marker = 'o', linestyle = '-.', linewidth = 2)
plt.title('Monthly GMV Year 2019', loc='center', fontsize = 20, color = 'blue', pad = 20)
plt.xlabel('Order Month',fontsize = 15)
plt.ylabel('Total GMV', fontsize =  15)
plt.show()
```

```python
# Kustomisasi Grid
plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color = 'green', marker = 'o', linestyle = '-.', linewidth = 2)
plt.title('Monthly GMV Year 2019', loc='center', fontsize = 20, color = 'blue', pad = 20)
plt.xlabel('Order Month',fontsize = 15)
plt.ylabel('Total GMV', fontsize =  15)
plt.grid(color = 'darkgray',linestyle = ':', linewidth = 1.5)
plt.show()
```

```python
# Kustomisasi Axis Ticks
plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color = 'green', marker = 'o', linestyle = '-.', linewidth = 2)
plt.title('Monthly GMV Year 2019', loc='center', fontsize = 20, color = 'blue', pad = 20)
plt.xlabel('Order Month',fontsize = 15)
plt.ylabel('Total Amount (Milyar)', fontsize =  15)
plt.grid(color = 'darkgray',linestyle = ':', linewidth = 1.5)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.show()
```

```python
# Menentukan Batas Minimum dan Maksimum Axis Ticks
plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color = 'green', marker = 'o', linestyle = '-.', linewidth = 2)
plt.title('Monthly GMV Year 2019', loc='center', fontsize = 20, color = 'blue', pad = 20)
plt.xlabel('Order Month',fontsize = 15)
plt.ylabel('Total Amount (Milyar)', fontsize =  15)
plt.grid(color = 'darkgray',linestyle = ':', linewidth = 1.5)
plt.ylim(ymin =  0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.show()
```

```python
# Menambahkan Informasi Pada Plot
fig = plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color = 'green', marker = 'o', linestyle = '-.', linewidth = 2)
plt.title('Monthly GMV Year 2019', loc='center', fontsize = 20, color = 'blue', pad = 20)
plt.xlabel('Order Month',fontsize = 15)
plt.ylabel('Total Amount (Milyar)', fontsize =  15)
plt.grid(color = 'darkgray',linestyle = ':', linewidth = 1.5)
plt.ylim(ymin =  0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.text(0.45, 0.72, 'The GMV Incresed Significantly on October 2019', transform = fig.transFigure, color = 'red')
plt.show()
```
