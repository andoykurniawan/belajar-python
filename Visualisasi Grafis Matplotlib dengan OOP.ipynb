```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,20, 0.5)
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8,0.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8,0.8])
axes.plot(x,y);
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8,0.8])
axes.plot(x,y);
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul Grafik');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3,0.3])
axes1.plot(x,y);
axes1.set_xlabel('Sumbu X Besar')
axes1.set_ylabel('Sumbu Y Besar')
axes1.set_title('Judul Grafik Utama');
axes2.plot(y,x);
axes2.set_xlabel('Sumbu X Kecil')
axes2.set_ylabel('Sumbu Y Kecil')
axes2.set_title('Judul Grafik Kedua');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3,0.3])
axes1.plot(x,y, color = 'red');
axes1.set_xlabel('Sumbu X Besar')
axes1.set_ylabel('Sumbu Y Besar')
axes1.set_title('Judul Grafik Utama');
axes2.plot(y,x, color = 'blue');
axes2.set_xlabel('Sumbu X Kecil')
axes2.set_ylabel('Sumbu Y Kecil')
axes2.set_title('Judul Grafik Kedua');
```
