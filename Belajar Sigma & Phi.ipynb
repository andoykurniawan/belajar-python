```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.plot(b1,'rs-')
plt.plot(np.cumsum(b1),'bo-')
plt.xlabel('Index')
plt.ylabel('Angka B')
plt.legend(['b1', 'Cummlative Sum'])
plt.show()
```

```python
plt.plot(b1,'rs-')
plt.plot(np.cumprod(b1),'bo-')
plt.xlabel('Index')
plt.ylabel('Angka B')
plt.legend(['b1', 'Cummlative Prod'])
plt.show()
```

```python
bil2 = np.arange(1,5)
```

```python
p1 = np.sum(bil2)
p2 = np.sum(bil2**2)

j1= p1/p2
j2= 1/p1
```

```python
print(f'Jawaban 1 hasilnya adalah {j1} \ jawaban 2 hasilnya adalah {j2}')
```
