```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(4,4), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,4), dpi=100)

axes[0].plot(x,y)
axes[0].set_title('Grafik 1')

axes[1].plot(x,y)
axes[1].set_title('Grafik 2')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,8), dpi=100)

axes[0].plot(x,y)
axes[0].set_title('Grafik 1')

axes[1].plot(x,y)
axes[1].set_title('Grafik 2')

plt.tight_layout()
plt.show()
```

# menambah legenda

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(x,x**3)
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 5')
ax.legend()
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 5')
ax.legend(loc=(0.2,0.2))
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 5')
ax.legend(loc=(0.5,0.5))
plt.show()
```

```python
fig = plt.figure(facecolor='yellow', dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 5')
ax.legend(loc=(0.2,0.2))
plt.show()
```

```python
fig = plt.figure(facecolor='yellow', dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')
ax.plot(x,x**4, label='Pangkat 4')
ax.legend(loc=(0.2,0.2))
plt.show()
```

```python
fig = plt.figure(facecolor='yellow', dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')
ax.plot(x,x**4, label='Pangkat 4')
ax.legend(loc=(0.5,0.5))
ax.grid(True)
plt.show()
```
