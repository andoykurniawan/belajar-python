```python
# versi 2 distribusi z (Normal) outlier
from scipy.stats import zscore
import numpy as np
```

```python
nilai_z = zscore(dataku.fare)
```

```python
nilai_z
```

```python
nilai_z_abs = np.abs(nilai_z)
```

```python
nilai_z_abs
```

```python
filter_data = (nilai_z_abs<3)
```

```python
dataku_baru = dataku[filter_data]
```

```python
dataku_baru
```

```python
sns.jointplot('age','fare', data = dataku_baru, kind='reg');
```

```python
sns.jointplot('age','fare', data = dataku_baru, kind='hex');
```

```python
sns.jointplot('age','fare', kind='kde', data = dataku_baru);
```
