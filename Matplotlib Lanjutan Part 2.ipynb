```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt

# dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/retail_raw_reduced.csv")
dataset = pd.read_csv("retail_raw_reduced.csv")

# untuk mendapatkan bulan dan tahun pada kolom order date
dataset['order_month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))

# Penambahan kolom GMV (Gross Merchandise Value) pada dataset / Nilai total
dataset['gmv'] = dataset['quantity'] * dataset['item_price']

# Buat Multi-LIne Chart
dataset.groupby(['order_month', 'brand'])['gmv'].sum().unstack().plot()

plt.title('Mothly GMV Year 2019 - Breakdown by Brand', loc = 'center', pad = 30, fontsize = 20, color = 'blue')
plt.xlabel('Order Month', fontsize = 15)
plt.ylabel('Total Amount (in Billions)', fontsize = 15)
plt.grid(color = 'darkgray', linestyle = ':', linewidth = 0.5)
plt.ylim(ymin = 0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))

plt.gcf().set_size_inches(10,5)
plt.tight_layout()

plt.show()
```

```python
# Kustomisasi Legend

dataset.groupby(['order_month', 'brand'])['gmv'].sum().unstack().plot()

plt.title('Mothly GMV Year 2019 - Breakdown by Brand', loc = 'center', pad = 30, fontsize = 20, color = 'blue')
plt.xlabel('Order Month', fontsize = 15)
plt.ylabel('Total Amount (in Billions)', fontsize = 15)
plt.grid(color = 'darkgray', linestyle = ':', linewidth = 0.5)
plt.ylim(ymin = 0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))

plt.gcf().set_size_inches(12,5)
plt.tight_layout()

plt.legend(loc = 'right', bbox_to_anchor = (1.3, 0.5), shadow = True, ncol = 2)

plt.show()
```

```python
# Kustomisasi ColorMap
dataset.groupby(['order_month', 'province'])['gmv'].sum().unstack().plot(cmap = 'Set1')

plt.title('Mothly GMV Year 2019 - Breakdown by Brand', loc = 'center', pad = 30, fontsize = 20, color = 'blue')
plt.xlabel('Order Month', fontsize = 15)
plt.ylabel('Total Amount (in Billions)', fontsize = 15)
plt.grid(color = 'darkgray', linestyle = ':', linewidth = 0.5)
plt.ylim(ymin = 0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))

plt.gcf().set_size_inches(10,5)
plt.tight_layout()

plt.legend(loc = 'lower center', bbox_to_anchor = (0.5, -0.5), shadow = True, ncol = 3, title = 'Provinsi', fontsize = 9, title_fontsize = 11)

plt.show()
```

```python
# Membuat variabel baru untuk 5 Provinsi dengan GMV tertinggi

top_provinces = dataset.groupby('province')['gmv'].sum().reset_index().sort_values('gmv', ascending = False).head(5)
top_provinces
```

```python
# Membuat Subset Data
dataset_dki_q4 = dataset[(dataset['province'] == 'DKI Jakarta') & (dataset['order_month'] >= '2019-10')]
dataset_dki_q4
    
```

```python
# Membuat Pie Chart
gmv_per_city_dki_q4 = dataset_dki_q4.groupby('city')['gmv'].sum().reset_index()
plt.figure(figsize = (6,6))
plt.pie(gmv_per_city_dki_q4['gmv'], labels = gmv_per_city_dki_q4['city'], autopct = '%1.2f%%')
plt.title('GMV Contribution Per City - DKI Jakarta in Q4 2019', loc = 'center', pad = 30, fontsize = 15, color ='blue')
plt.show()
```

```python
# Membuat Bar Chart
dataset_dki_q4.groupby('city')['gmv'].sum().sort_values(ascending = False).plot(kind = 'bar', color = 'green')
plt.title('GMV Contribution Per City - DKI Jakarta in Q4 2019', loc = 'center', pad = 30, fontsize = 15, color ='blue')
plt.xlabel('City', fontsize = 15)
plt.ylabel('Total Amount(in Billion)', fontsize = 15)
plt.ylim(ymin = 0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.xticks(rotation = 0)
plt.show()
```

```python
# Membuat Multi Bar Chart
dataset_dki_q4.groupby(['city', 'order_month'])['gmv'].sum().unstack().plot(kind = 'bar')
plt.title('GMV Contribution Per City - DKI Jakarta in Q4 2019', loc = 'center', pad = 30, fontsize = 15, color ='blue')
plt.xlabel('Province', fontsize = 15)
plt.ylabel('Total Amount(in Billion)', fontsize = 15)
plt.legend(bbox_to_anchor = (1,1), shadow = True, title = 'Month')
plt.ylim(ymin = 0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Membuat Stacked Chart
dataset_dki_q4.groupby(['order_month','city'])['gmv'].sum().sort_values(ascending = False).unstack().plot(kind = 'bar', stacked = True)
plt.title('GMV Contribution Per City - DKI Jakarta in Q4 2019', loc = 'center', pad = 30, fontsize = 15, color ='blue')
plt.xlabel('Order Month', fontsize = 12)
plt.ylabel('Total Amount(in Billion)', fontsize = 12)
plt.legend(bbox_to_anchor = (1,1), shadow = True, title = 'Month')
plt.ylim(ymin = 0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()
```

```python
# Membuat Agregat Data Customer
data_per_customer = (dataset_dki_q4.groupby('customer_id').agg({'order_id':'nunique', 'quantity':'sum', 'gmv':'sum'}).reset_index().rename(columns={'order_id':'orders'}))
data_per_customer.sort_values(by = 'orders', ascending = False)
```

```python
# Membuat Histogram Part 1
plt.clf()

# histogram pertama
plt.figure()
plt.hist(data_per_customer['orders'])
plt.show()

# histogram kedua
plt.figure()
plt.hist(data_per_customer['orders'], range = (1,5))
plt.title('Distribution of Number of Orders per Customer\nDKI Jakarta in Q4 2019', fontsize = 15, color = 'blue')
plt.xlabel('Number of Orders', fontsize = 12)
plt.ylabel('Number of Customers', fontsize = 12)
plt.show()
```

```python
# Membuat Histogram Part 2
plt.figure(figsize=(10,5))
plt.hist(data_per_customer['quantity'], bins = 100, range = (1, 200), color = 'brown')
plt.title('Distribution of Total Quantity Customer\nDKI Jakarta in Q4 2019', fontsize = 15, color = 'blue')
plt.xlabel('Quantity', fontsize = 12)
plt.ylabel('Number of Customers', fontsize = 12)
plt.xlim(xmin = 0, xmax = 200)
plt.show()
```

```python
# Membuat Histogram Part 3
plt.figure(figsize=(10,5))
plt.hist(data_per_customer['gmv'], bins = 100, range = (1, 200000000), color = 'green')
plt.title('Distribution of Total GMV Customer\nDKI Jakarta in Q4 2019', fontsize = 15, color = 'blue')
plt.xlabel('GMV (in Millions)', fontsize = 12)
plt.ylabel('Number of Customers', fontsize = 12)
plt.xlim(xmin = 0, xmax = 200000000)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.show()
```

```python
# Membuat Scatter Plot
plt.clf()

# histogram pertama
plt.figure()
plt.scatter(data_per_customer['quantity'], data_per_customer['gmv'])
plt.show()

# histogram kedua
plt.figure()
plt.scatter(data_per_customer['quantity'], data_per_customer['gmv'], marker = '+', color = 'red')
plt.title('Correlation of Quantity and GMV per Customer\nDKI Jakarta in Q4 2019', fontsize = 15, color = 'blue')
plt.xlabel('Quantity', fontsize = 12)
plt.ylabel('GMV (in Billions)', fontsize = 12)
plt.xlim(xmin = 0, xmax = 300)
plt.ylim(ymin = 0, ymax = 150000000)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.show()
```
