```python
sns.PairGrid(berlian)
```

```python
p = sns.PairGrid(berlian)
p.map(plt.scatter)
plt.show()
```

```python
p = sns.PairGrid(berlian)
p.map_diag(sns.displot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
plt.show()
```

```python
tabrak = sns.load_dataset('car_crashes')
```

```python
p = sns.PairGrid(tabrak)
p.map_diag(sns.displot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
p = sns.FacetGrid(tips, row='smoker', col='day')
```

```python
p = sns.FacetGrid(tips, row='smoker', col='day')
p.map(sns.distplot, 'total_bill')
plt.show()
```

```python
p = sns.FacetGrid(tips, row='smoker', col='time')
p.map(sns.distplot, 'total_bill')
plt.show()
```

```python
p = sns.FacetGrid(tips, row='smoker', col='time')
p.map(sns.distplot, 'total_bill', kde=False)
plt.show()
```

```python
p = sns.FacetGrid(tips, row='smoker', col='time')
p.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
p = sns.FacetGrid(tips, row='smoker', col='time', hue='sex')
p.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```
